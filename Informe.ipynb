# **GENERADOR DE OFERTAS DE PRODUCTOS**

## **PROPÓSITO DEL PROYECTO**
El propósito principal de este proyecto fue desarrollar un sistema ETL (Extract, Transform, Load) con Apache Airflow para automatizar la generación de ofertas un producto, para este proyecto se usó como ejemplo la API de Mercado Libre y Celulares. El objetivo era identificar productos con descuentos notables en la API de Mercado Libre y enviar ofertas por correo electrónico a los clientes.

## **DETALLES DEL PROYECTO**

### **TECNOLOGÍAS UTILIZADAS**




*   Apache Airflow: Utilizado como el motor de ejecución de tareas y programación de trabajos para automatizar flujos de trabajo.
*   Docker: Empleado para crear contenedores de aplicaciones y orquestar el entorno de ejecución de Apache Airflow de manera eficiente y portátil.


*   Visual Studio Code (VSCode): Utilizado como entorno de desarrollo para escribir, depurar y administrar código.
*   Postman: Herramienta utilizada para probar y depurar APIs.







### **DESCRIPCIÓN BREVE DE LOS SCRIPTS UTILIZADOS**
**generador_ofertas.py**

Aquí se define el flujo de trabajo DAG en Apache Airflow que consta de dos tareas:


*   task_1(Consultando_API): Esta tarea consulta la API de Mercado Libre para obtener información sobre productos de la categoría "MLA1055", que corresponde a la categoría “Celulares y Smartphones”. La tarea 1 llama a la función “consultar_api” que a su vez ésta llama a la función “get_most_relevant_items_for_category” encargada de devolver una lista de diccionarios con la información relevante.
*   task_2(Generando_oferta): Esta tarea utiliza el resultado de la consulta a la API para analizar y generar ofertas personalizadas. Accede al resultado de la Tarea 1 utilizando XCom y luego ejecuta la función “analizar_y_ofertar” importado del script ofertar.py para generar ofertas.

**consultar_api1.py**

Se encarga de consultar la API de Mercado Libre y guardar información relevante en una lista de diccionarios. La API se consulta para obtener los productos más relevantes de una categoría específica, y luego se filtran y almacenan los datos esenciales.


**ofertar.py**

Se encarga de analizar los datos recopilados previamente, identificar productos con descuentos notables y enviar ofertas por correo electrónico a los clientes interesados. Los datos de productos provienen de la API de Mercado Libre y se procesan para encontrar ofertas atractivas.
